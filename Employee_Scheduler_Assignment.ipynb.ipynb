```python
import random


DAYS = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
SHIFTS = ["Morning", "Afternoon", "Evening"]


class Employee:
    def __init__(self, name, preferences):
        self.name = name
        self.preferences = preferences
        self.assigned_days = 0



def create_schedule(employees):

    # Creating empty schedule structure
    schedule = {
        day: {shift: [] for shift in SHIFTS}
        for day in DAYS
    }


    for day in DAYS:
        for emp in employees:

            if emp.assigned_days >= 5:
                continue

            preferred_shift = emp.preferences.get(day)

            # Checking shift capacity and one shift per day rule
            if (len(schedule[day][preferred_shift]) < 2 and
                not any(emp.name in schedule[day][s] for s in SHIFTS)):

                schedule[day][preferred_shift].append(emp.name)
                emp.assigned_days += 1

   
    for day in DAYS:
        for shift in SHIFTS:

            while len(schedule[day][shift]) < 2:

                available = []

                for emp in employees:

                    if emp.assigned_days >= 5:
                        continue

                    # Ensuring employee not already assigned that day
                    if any(emp.name in schedule[day][s] for s in SHIFTS):
                        continue

                    available.append(emp)

                if not available:
                    break

                chosen = random.choice(available)
                schedule[day][shift].append(chosen.name)
                chosen.assigned_days += 1

    return schedule



def print_schedule(schedule):

    print("\nFINAL WEEKLY SCHEDULE\n")

    for day in DAYS:
        print(f"--- {day} ---")
        for shift in SHIFTS:
            print(f"{shift}: {', '.join(schedule[day][shift])}")
        print()


-

employees = [
    Employee("Alice", {day: "Morning" for day in DAYS}),
    Employee("Bob", {day: "Afternoon" for day in DAYS}),
    Employee("Charlie", {day: "Evening" for day in DAYS}),
    Employee("David", {day: "Morning" for day in DAYS}),
    Employee("Eva", {day: "Afternoon" for day in DAYS}),
    Employee("Frank", {day: "Evening" for day in DAYS}),
    Employee("Grace", {day: "Morning" for day in DAYS}),
    Employee("Henry", {day: "Afternoon" for day in DAYS}),
    Employee("Ivy", {day: "Evening" for day in DAYS}),
]

schedule = create_schedule(employees)
print_schedule(schedule)
```
